```python
# Creating a DataFrame consisting only of numerical values and targets
correlation_df = df.copy()

# Convert only target columns to binary format
drug_columns = [
   'Alcohol', 'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc',
    'Coke', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'Legalh', 'LSD',
    'Meth', 'Mushrooms', 'Nicotine', 'Semer', 'VSA'
]

for col in drug_columns:
    # Set the value of ‘CL0’ to 0, set the others to 1
    correlation_df[col] = correlation_df[col].apply(lambda x: 0 if x == 'CL0' else 1)

# Creating a correlation matrix between numerical and target variables
all_features = numerical_features + drug_columns
corr_matrix = correlation_df[all_features].corr()

# Visualizing the correlation matrix with a heatmap
plt.figure(figsize=(15,9))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Matrix: Personality Traits and Substance Use")
plt.show()
```
